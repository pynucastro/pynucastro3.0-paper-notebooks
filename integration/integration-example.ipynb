# Example of integrating

```python
import pynucastro as pyna
```

```python
net = pyna.network_helper(["h1", "h2", "he3", "he4", "c12", "c13", "n13", "n14", "n15", "o16"])
```

```python
rho = 150
T = 1.5e7
comp = pyna.Composition(net.unique_nuclei, init="solar")
```

```python
comp.get_sum_X()
```

```python
tmax = 1.e20
sol = net.integrate_network(tmax, rho, T, comp, atol=1.e-8)
```

```python
fig = sol.plot_evolution(ymin=1.e-8, tmin=1.e6)
```

```python
fig.set_size_inches(7, 5)
fig.tight_layout()
fig.savefig("integration-example.pdf", bbox_inches="tight")
```
